```python
import os
from datetime import datetime

# Folder path
folder_path = r"C:\Users\user\Downloads\27-10-25 1\27-10-25\XML - 2025-10-27"

search_text = 'folder="2422485"'
matched_files = []

for file_name in os.listdir(folder_path):
    if file_name.lower().endswith(".xml"):
        file_path = os.path.join(folder_path, file_name)

        try:
            with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
                content = f.read()
                if search_text in content:
                    modified_time = os.path.getmtime(file_path)
                    modified_datetime = datetime.fromtimestamp(modified_time)

                    matched_files.append((file_name, modified_datetime))
        except Exception as e:
            print(f"Error reading {file_name}: {e}")

# Output results
print("Files containing folder=\"2422396\":\n")

for file, mod_time in matched_files:
    print(f"File: {file}")
    print(f"Modified Date & Time: {mod_time}")
    print("-" * 50)

print(f"\nTotal files found: {len(matched_files)}")
```

```python
import zipfile
from datetime import datetime

# Actual zip file path (ONLY up to .zip)
zip_path = r"C:\Users\user\Downloads\27-10-25 1.zip"

# Folder INSIDE the zip
inner_folder = "27-10-25/XML - 2025-10-27/"

search_text = 'folder="2422485"'
matched_files = []

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    for file_name in zip_ref.namelist():

        # Only XML files inside required folder
        if file_name.startswith(inner_folder) and file_name.lower().endswith(".xml"):
            with zip_ref.open(file_name) as f:
                content = f.read().decode("utf-8", errors="ignore")

                if search_text in content:
                    info = zip_ref.getinfo(file_name)
                    modified_datetime = datetime(*info.date_time)

                    matched_files.append((file_name, modified_datetime))

# Output
print("\nXML files containing folder=\"2422485\":\n")

for file, mod_time in matched_files:
    print(f"XML File      : {file}")
    print(f"Modified Time : {mod_time}")
    print("-" * 60)

print(f"\nTotal XML files found: {len(matched_files)}")
```
